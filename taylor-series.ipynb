# Aproximações para π

## Fórmula de MACH

## Série de Taylor

arctan(x) = (∞∑0 (-1)^n) . (x^(2n+1)/(2n+1))

Construindo um programa que permita estimar o valor de π a partir do número de termos da série de Taylor.

```python
def taylor(t_precision):
    t_list = []
    t_number = 0
    t_final = 0
    teste = 0
    while teste < t_precision:
        
        t_answer = (-1)**t_number / (2*t_number+1)
        print(t_answer)

        t_list.append(t_answer)
        t_final = 4 * (sum(t_list))
        print(t_final)
        t_number = t_number + 1
        teste = teste + 1


    return (t_final,t_number)

t_precision = 1000
taylor(t_precision)
```
